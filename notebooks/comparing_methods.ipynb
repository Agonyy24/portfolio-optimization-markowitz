## Comparing Optimization Methods: Matrix Formula vs. SciPy Minimize

Both of the methods lead to very similar results of the portfolio details. Small differences are caused by numerical rounding errors and the iterative approach in scipy.optimize.minimize vs. the closed-form matrix solution.


## Tangency Portfolio and Risk-Free Rate

Large discrepancies appeared initially because the matrix-based version assumed Risk free rate = 0.

## Differences between methods

The matrix formula assumes no constraints on weights (they can be negative or exceed 1).

scipy.optimize.minimize makes it easy to add bounds (e.g., (0,1) or (-1,1)) and additional constraints.

Results should only be compared when both methods use the same assumptions.

## Conclusion

Both methods converge to nearly identical results under consistent assumptions, but serve different purposes: the matrix formula is perfect for theory and fast validation, while numerical optimization is the go-to for real-world constraints.
