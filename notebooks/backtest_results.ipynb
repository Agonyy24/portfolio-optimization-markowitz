# Backtest Results - MVP vs. Tangency Portfolio
## Key Observations

### Tangency Portfolio (TP):

Reached a peak cumulative growth of around 2.0x, showing strong performance during bullish periods.

However, it experienced much higher volatility, with sharp drawdowns during market pullbacks.

This behavior is expected - TP maximizes the Sharpe ratio, which usually implies a more aggressive allocation.

### Minimum Variance Portfolio (MVP):

Grew steadily throughout the year, ending at about 1.3x cumulative growth.

Displayed much smoother performance with smaller fluctuations.

This makes it a more defensive and lower-risk choice compared to TP.

# Conclusions

According to Modern Portfolio Theory, MVP represents the combination of stocks with the lowest volotality, which appeals risk-averse investors. However, when we expand MPT with Capital Market Line (CML) from CAPM theory, it becomes suboptimal. This is because an investor can achieve the same level of return with even lower volatility (or a higher return for the same risk) by combining the Tangency Portfolio with a risk-free asset.

The mathematical "best" basket of risky assets-the Tangency Portfolio (TP) stays the same for everyone, regardless of their age or goals, because it offers the highest Sharpe ratio. Once this ideal mix is identified, the only thing that changes is how much risk an investor wants to take. By moving along the Capital Market Line (CML), a cautious investor can mix the TP with risk-free assets (i.e. bonds), while an aggressive investor can use leverage to buy more of it. 

While MPT and CAPM provide a solid theoretical foundation, they do not explain everything in today’s complex markets. These models assume that investors are always rational and that returns follow a normal distribution, which is often not the case during black swan events or extreme volatility. This is where Quants (hopefully me one day 😂) take over, using advanced algorithms, alternative data, and machine learning to move beyond simple linear relationships. While the Tangency Portfolio is a great mathematical starting point, modern Quants constantly refine these strategies to account for market anomalies that classical theory simply cannot capture.
