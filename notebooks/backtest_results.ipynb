# Backtest Results – MVP vs. Tangency Portfolio (2024)
## Key Observations

### Tangency Portfolio (TP):

Reached a peak cumulative growth of around 2.0×, showing strong performance during bullish periods.

However, it experienced much higher volatility, with sharp drawdowns during market pullbacks.

This behavior is expected – TP maximizes the Sharpe ratio, which usually implies a more aggressive allocation.

### Minimum Variance Portfolio (MVP):

Grew steadily throughout the year, ending at about 1.3× cumulative growth.

Displayed much smoother performance with smaller fluctuations.

This makes it a more defensive and lower-risk choice compared to TP.

![Backtest](../plots/backtest.png)

# Conclusions

The TP outperformed MVP in absolute return but with significantly higher risk.

Risk-adjusted performance (Sharpe ratio) would likely still favor TP, but it might not suit conservative investors.

MVP demonstrated its purpose: stability and capital preservation, even if it sacrifices some upside potential.

### In practice, a combination of MVP and TP could be used to create a portfolio that balances growth potential with risk tolerance.
